```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
sw = nlp.Defaults.stop_words
print(sw)
```

```python
print(len(sw))
```

```python
print(nlp.vocab['I'].is_stop)
```

```python
print(nlp.vocab['monjir'].is_stop)
```

```python
def add_to_stop(word):
    nlp.Defaults.stop_words.add(str(word))
    nlp.vocab[str(word)].is_stop = True
```

```python
add_to_stop('moti')
```

```python
print(nlp.vocab['moti'].is_stop)
```

```python
nlp.Defaults.stop_words
```

```python
def remove_from_stop(word):
    nlp.Defaults.stop_words.remove(str(word))
    nlp.vocab[str(word)].is_stop = False
```

```python
remove_from_stop('moti')
```

```python
print(nlp.vocab['moti'].is_stop)
```
